```python
from calendar import month_name

def sort_month(df, colname):
    # Create a dictionary to map month names to their respective numbers (skip the empty string)
    month_dict = {j: i for i, j in enumerate(month_name) if j}  # month_name starts with an empty string
    
    # Apply the mapping based on month names
    df["month_num"] = df[colname].apply(lambda x: month_dict.get(x, 0))  # Use get to avoid KeyError
    
    # Sort by the 'month_num' column and clean up
    return df.sort_values(by="month_num").reset_index(drop=True).drop(['month_num'], axis=1)
```

```python
# Assuming 'cancellation' is already added to data_cat DataFrame
data_cat['cancellation'] = final_data['is_canceled']

# Define the mean encoding function
def mean_encode(df, col, mean_col):
    # Create the dictionary based on the group mean of the categorical variable
    df_dict = df.groupby([col])[mean_col].mean().to_dict()
    
    # Map the dictionary to the column
    df[col] = df[col].map(df_dict)
    
    # Handle categories that might not be present in the map
    df[col].fillna(df[mean_col].mean(), inplace=True)
    
    return df

# List of categorical columns you want to mean encode
categorical_columns = ['hotel', 'arrival_date_month', 'meal', 'market_segment', 
                       'distribution_channel', 'reserved_room_type', 'assigned_room_type', 
                       'deposit_type', 'customer_type']

# Apply mean encoding to each categorical column
for col in categorical_columns:
    data_cat = mean_encode(data_cat, col, 'cancellation')

# Drop the 'cancellation' column after encoding
data_cat.drop(["cancellation"], axis=1, inplace=True)

# Output the encoded data_cat
print(data_cat.head())
```
